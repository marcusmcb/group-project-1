```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pop_growth_df = pd.read_csv('us_population_data.csv', encoding="ISO-8859-1")
pop_growth_df.head()
```

```python
population_growth = pop_growth_df.iloc[3:,8:].median()
population_growth
```

```python

x_axis = population_growth.index
plt.bar(x_axis, population_growth, color='r', alpha=0.5, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017"])
plt.ylim(200000, max(population_growth)+1000)
plt.title("Average Annual US Population Growth - Top 400 Market")
plt.ylabel("Population")
plt.xlabel("source: US Census Data (www.census.gov)")
```
